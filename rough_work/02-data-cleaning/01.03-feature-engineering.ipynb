#01.02 Feature Engineering
Creating new features to improve predictions for the Digital Green Crop Yield Estimate Challenge.

# Prepare Workspace

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import warnings
warnings.filterwarnings("ignore")

```

```python
data = pd.read_csv('cleaned_fulldf.csv')
```

```python
data.head()
```

### Perform Feature Engineering

```python
# Create feature for past month yield per acre 'Past_Yield_per_Acre'
data['Harv_date'] = pd.to_datetime(data['Harv_date'])
data.sort_values(['District', 'Harv_date'], inplace=True)

# Group the DataFrame by 'District' and calculate the rolling average
data['Past_YpA_Avg'] = data.groupby('District')['Yield_per_Acre'].rolling(
    window = 30).mean().reset_index(0, drop=True)

# Fill NaN values in the 'past_month_avg' column with 0 if needed
data['Past_YpA_Avg'].fillna(0, inplace=True)
```

```python
# Create feature for days between harvesting and threshing 'Days_Harv_Thresh'
data['Threshing_date'] = pd.to_datetime(data['Threshing_date'])
data['Days_Harv_Thresh'] = (
    data['Threshing_date'] - data['Harv_date']).dt.days
```
